```python
from tqdm import tqdm
def load_endomondo_sample(filepath, n_samples=1000):
    """Load first n_samples from Endomondo JSON file."""
    data = []
    
    # Créer une barre de progression
    with open(filepath, 'r') as f:
        for i in tqdm(range(n_samples), desc="Loading workouts"):
            line = f.readline()
            if not line:
                break
            try:
                workout = ast.literal_eval(line.strip())
                data.append(workout)
            except:
                continue
    
    return data
```

```python
# Load sample
n_samples=50000
endomondoHR_df = load_endomondo_sample(endomondoHR_path, n_samples=n_samples)
print(f"Loaded {len(endomondoHR_df)} workouts")

endomondoHR_df = pd.DataFrame(endomondoHR_df)
endomondoHR_df = endomondoHR_df[endomondoHR_df['sport'] == 'run']

cols_to_check = ['speed','altitude','gender','heart_rate','userId']
endomondoHR_df = endomondoHR_df.dropna(subset = cols_to_check)
endomondoHR_df.count()
```
